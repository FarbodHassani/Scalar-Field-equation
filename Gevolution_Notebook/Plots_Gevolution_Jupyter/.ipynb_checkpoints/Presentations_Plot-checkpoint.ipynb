# Background and cosmological parameters

```python
#################################
#################################
# Cosmological parameters.
As=2.215*1.e-9;
h=0.67556;
kp=0.05/h; 
ns=0.9619;
cs2_e3=1.e-6;
c=2.99792458*1.e5; #Km/s
# Boxsize=320.;
# c_gev=2997.92458; #[100km/s]
# H0_conf_Gev=np.sqrt(Boxsize**2/c**2); #0.10677 rho_crit in Gev=1
H0=100*h/c; #H0 in 1/Mpc unit. 0.00022593979933110373
# print("H0[1/Mpc]=100h/c: ",H0);
w=-0.9;
print("H0[1/Mpc]=100h/c: ",H0);
def Hubble_conf_Mpc(a):
    c=2.99792458*1.e5;h=0.67556;H0=100*h/c;w=-0.9;
    Omega_b=0.022032/h/h; Omega_cdm=0.12038/h/h;
    Omega_m=Omega_b+Omega_cdm; Omega_Lambda=0;
    Omega_rad=9.1671353942930788e-05; Omega_kessence=1.-Omega_m-Omega_rad;
    return H0*np.sqrt(Omega_m*(a**-3)+Omega_rad*(a**-4)+Omega_kessence*(a**(-3*(1+w))))*a
Omega_b=0.022032/h/h; Omega_cdm=0.12038/h/h;
Omega_m=Omega_b + Omega_cdm;
z_all=[99.597220, 48.322258, 9.951253, 5.808126, 0.988975, 0.]
z_list=[100, 50, 10, 6, 1, 0]
# List of scale factor
a100=(1./(1.+z_all[0])); a50=(1./(1.+z_all[1])); a10=(1./(1.+z_all[2])); a6=(1./(1.+z_all[3]));
a1=(1./(1.+z_all[4])); a0=(1./(1.+z_all[5]));
a_all=[a100,a50,a10,a6,a1,a0]
H_conf_z100=Hubble_conf_Mpc(1./(1.+z_all[0])); # Unit=1/Mpc Hconf=a*H
H_conf_z50=Hubble_conf_Mpc(1./(1.+z_all[1])); # Unit=1/Mpc Hconf=a*H
H_conf_z10=Hubble_conf_Mpc(1./(1.+z_all[2])); # Unit=1/Mpc Hconf=a*H
H_conf_z6=Hubble_conf_Mpc(1./(1.+z_all[3])); # Unit=1/Mpc Hconf=a*H
H_conf_z1=Hubble_conf_Mpc(1./(1.+z_all[4])); # Unit=1/Mpc Hconf=a*H
H_conf_z0=Hubble_conf_Mpc(1./(1.+z_all[5])); # Unit=1/Mpc Hconf=a*H
H_conf_zall_list=[H_conf_z100, H_conf_z50, H_conf_z10,
                  H_conf_z6, H_conf_z1, H_conf_z0];
```
