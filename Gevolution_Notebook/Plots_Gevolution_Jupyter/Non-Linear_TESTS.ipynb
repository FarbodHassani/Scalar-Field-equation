# Notebook for comparison beween  Gevolution and class and agreement between class and hi-class

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.interpolate import interp1d
from pylab import rcParams
import scipy.integrate as integrate
ColorsI = ["red","blue","green","purple","darkcyan","gold","maroon","black","darkslategrey",
           "darkgreen","navy","indigo","magenta","salmon","dodgerblue","springgreen","plum",
           "cyan","khaki","lightcoral","gray"]
ColorsII = []

# plt.style.use('seaborn-paper')
plt.rcParams['font.serif'] = 'Ubuntu'
```

```python
addressGev="./../../../Kessence-NLTest/output/";
Gev_nl_zeta_zall=[]
Gev_nl_Hpi_zall=[]
Gev_nl_phi_zall=[]
z_list=[ 100.0 ,10,5, 1.,0.5, 0.3, 0.0,-0.4,-0.85,-0.89,-0.893,-0.897,-0.896,-0.899,-0.9]
maxNum=np.shape(z_list)[0]
for i in range (0,maxNum):
    Gev_nl_zeta_zall.append(np.loadtxt("./"+addressGev+"/kess_pk_cs_0_"+str(i).zfill(3)+"_zeta.dat"))
    Gev_nl_Hpi_zall.append(np.loadtxt("./"+addressGev+"/kess_pk_cs_0_"+str(i).zfill(3)+"_pi_k.dat"))
    Gev_nl_phi_zall.append(np.loadtxt("./"+addressGev+"/kess_pk_cs_0_"+str(i).zfill(3)+"_phi.dat") )
average=np.loadtxt("./"+addressGev+"/Result_avg.txt")
```

# Average Plots

```python
plt.figure(figsize=(14,8))
ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 18)
plt.loglog(average[:,0],average[:,1],'r.',lw=2.5,label=r"$\langle \pi \rangle$")
plt.loglog(average[:,0],average[:,2],'b.',lw=2.5,label=r"$\langle \zeta \rangle$")
plt.loglog(average[:,0],average[:,3],'g.',lw=2.5,label=r"$\langle \Phi \rangle$")


plt.ylabel(r"$\langle f \rangle $",fontsize=18)
plt.xlabel(r"$\frac{\tau}{{Boxsize}}$",fontsize=18)

# plt.ylabel('log')
# plt.xlabel('log')
# plt.grid(True)
plt.legend(bbox_to_anchor=(.2, 0.40, 0.4, .102), loc=1,ncol=1,fontsize=12, mode="expand", borderaxespad=0.)

# ax1.set_xticklabels([]) 
plt.show()
```
