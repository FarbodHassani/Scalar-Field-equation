```python
import numpy as np
import matplotlib.pylab as plt
from scipy.interpolate import interp1d
```

# Stress tensor powerspectrum

```python
backgr=np.loadtxt("./lcdm_background.dat")
```

```python
a=backgr[:,2]
h=0.67556
Omega_m=0.312046;
Omega_kessence=0.67;
w_kess=-0.9;
Omega_L=(1-Omega_m-Omega_kessence);
myHa=np.sqrt(Omega_m/a+(Omega_kessence)*a**(-3*(1+w_kess))*a*a+Omega_L*a*a)
plt.loglog(backgr[:,2],backgr[:,3],color="blue")
plt.loglog(a,myHa,color="red")
# plt.loglog(bgnew[:,2],bgnew[:,3],color="Black")
plt.xlim(0.008,1.05)
#plt.ylim(1.e-17,1.e-1)
plt.show()
```
